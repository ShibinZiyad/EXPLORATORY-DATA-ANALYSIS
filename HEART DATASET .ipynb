```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
  
df = pd.read_csv('heart.csv')
df.head()
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.displot(df['Age'], kde = True, color = 'blue')
plt.xlabel("Age")
plt.ylabel("Density")
plt.title("Distribution of Age")

plt.show()
```

```python
sns.displot(x = df['RestingBP'], kde = True, color = 'blue')
plt.xlabel("RestingBP")
plt.ylabel("Density")
plt.title("Distribution of RestingBP")
plt.show()
```

```python
sns.displot(df['Cholesterol'], kde = True, color = 'blue')
```

```python
sns.displot(df['MaxHR'], kde = True, color ="blue")
```

```python
sns.violinplot(x = df['Age'])
plt.xlabel("Age")
plt.ylabel("Density")
plt.title("Distribution of Age")
plt.show()
```
